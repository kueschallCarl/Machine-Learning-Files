## Aufgabe 2: Binäre Klassifikation mit Logistischer Regression

In dieser Aufgabe soll ein binäres Klassifikationsmodell entwickelt werden, um vorherzusagen, ob ein Patient an Diabetes leidet oder nicht.

#### a) Daten vorbereiten

Lesen Sie den Diabetes-Datensatz ein. Spalten Sie die Zielvariable (Outcome) ab und splitten Sie die Daten anschließend in Trainings- und Testdaten, sodass die Testdaten 30% der ursprünglichen Samples enthalen.

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
X = df[['Pregnancies', 'Glucose','BloodPressure','SkinThickness', 'Insulin', 'BMI',
        'DiabetesPedigreeFunction','Age']]
y = df['Outcome']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=55)
```

#### b) Logistisches Regressionsmodell auf dem Datensatz trainieren

Trainieren Sie ein Logistisches Regressionsmodell auf den Trainingsdaten.

```python
#Scaling
#Scale X
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

#Fit model
model = LogisticRegression(max_iter=1000)
model_scaled = model.fit(X_train_scaled, y_train)

#Scale Test with same Scaler
X_test_scaled = scaler.transform(X_test)

#predict probabilities
y_pred_train_scaled_proba = model_scaled.predict_proba(X_train_scaled)[:, 1]
y_pred_test_scaled_proba = model_scaled.predict_proba(X_test_scaled)[:, 1]
```

Calculating the best performing Threshold:

Scaling the data massively improved the models performance in this case!!!

```python
# Calculate ROC-AUC scores for different thresholds
thresholds = np.linspace(0, 1, num=100)  # Generate a range of thresholds
roc_scores = []
for threshold in thresholds:
    y_pred = (y_pred_train_scaled_proba >= threshold).astype(int)
    roc_score = roc_auc_score(y_train, y_pred)
    roc_scores.append(roc_score)

# Find the threshold with the highest ROC score

best_threshold = thresholds[np.argmax(roc_scores)]
best_roc_score = np.max(roc_scores)

print("Best Threshold:", best_threshold)
print("Highest ROC Score:", best_roc_score)
```

using that threshold moving forward:

```python
y_pred_train_thresholded = (y_pred_train_scaled_proba >= best_threshold).astype(int)
y_pred_test_thresholded = (y_pred_test_scaled_proba >= best_threshold).astype(int)
```

#### c) Modellanwendung -und -bewertung

Wenden Sie das Modell auf den Trainingsdatensatz und auf den Testdatensatz an und berechnen Sie jeweils die Konfusionsmatrix, die Accuracy, die Precision und den Recall. Verwenden Sie dazu die Funktionen `confusion_matrix`, `accuracy_score`, `precision_score` und `recall_score` aus dem Modul `sklearn.metrics`.

Standardmäßig wird bei der Logistischen Regression ein Sample als positiv klassifiziert, wenn der Schwellwert 0.5 überschritten wird. Verwenden Sie nun den Schwellwert 0.8 und berechnen Sie erneut Precision und Recall.<br><br>
Hinweis: die durch die Logistische Regression ermittelten Scores erhalten Sie durch Aufruf der Methode `predict_proba`.

```python
# Calculate the confusion matrix
cm = confusion_matrix(y_train, y_pred_train_thresholded)
print("Confusion Matrix:")
print(cm)

# Calculate the accuracy score
accuracy = accuracy_score(y_train, y_pred_train_thresholded)
print("Accuracy:", accuracy)

# Calculate the precision score
precision = precision_score(y_train, y_pred_train_thresholded)
print("Precision:", precision)

# Calculate the recall score
recall = recall_score(y_train, y_pred_train_thresholded)
print("Recall:", recall)
```

#### e) ROC-Kurve zeichnen

Berechnen Sie die ROC-AUC-Scores für den Trainings- und den Testdatensatz und zeichen Sie die ROC-Kurven für den Trainings- und den Testdatensatz in ein gemeinsames Diagramm.

np.argmax(roc_scores): The argmax function from the NumPy library returns the index of the maximum value in the roc_scores list. It finds the index corresponding to the threshold that resulted in the highest ROC score.

```python
# Calculate ROC-AUC score
#Training Data
roc_auc_train = roc_auc_score(y_train, y_pred_train_thresholded) 
fpr_train, tpr_train, thresholds_train = roc_curve(y_train, y_pred_train_thresholded) 

#Test Data
roc_auc_test = roc_auc_score(y_test, y_pred_test_thresholded) 
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_pred_test_thresholded)  

# Visualize the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, label='ROC curve Train(AUC = %0.2f)' % roc_auc_train)
plt.plot(fpr_test, tpr_test, label='ROC curve Test(AUC = %0.2f)' % roc_auc_test)
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line representing random guessing
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR) = Recall')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

#### f) Precision-Recall-Kurve zeichnen

Zeichen Sie die Precision-Recall-Kurve für den Trainings- und den Testdatensatz in ein gemeinsames Diagramm.

```python
precision_train, recall_train, thresholds_train = precision_recall_curve(y_train, y_pred_train_thresholded)
precision_test, recall_test, thresholds_test = precision_recall_curve(y_test, y_pred_test_thresholded)

plt.figure(figsize=(8, 6))
plt.plot(recall_train, precision_train, label='Prec-Rec-Curve Training Data')
plt.plot(recall_test, precision_test, label='Prec-Rec-Curve Test Data')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower right")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()
```

#### c) Modelltraining

Trainieren Sie auf dem Trainingsdatensatz ein Logistisches Regressionsmodell unter Verwendung eines multinomialen Ansatzes. Standardisieren Sie die Trainingsdaten vorher, indem Sie den `StandardScaler` anwenden.

Der multinomiale Ansatz bei der Logistischen Regression wird verwendet, wenn wir mehr als zwei Klassen in unserem Zielvariablenvektor haben. Im Gegensatz zur binären Logistischen Regression, bei der nur zwei Klassen vorhanden sind, können beim multinomialen Ansatz mehrere Klassen vorliegen.

Bei der normalen Logistischen Regression (auch binäre Logistische Regression genannt) wird eine Sigmoid-Funktion verwendet, um die Wahrscheinlichkeit einer binären Klassenzugehörigkeit (z. B. ja/nein, 0/1) zu schätzen. Das Modell lernt die Koeffizienten (Gewichte) für die Features und passt sie an, um die beste Trennung zwischen den beiden Klassen zu finden.

Beim multinomialen Ansatz geht es darum, ein Modell zu entwickeln, das für mehrere Klassen geeignet ist. Hier wird eine sogenannte Softmax-Funktion verwendet, die eine Wahrscheinlichkeitsverteilung über alle Klassen liefert. Das Modell schätzt die Wahrscheinlichkeiten für jede Klasse und wählt die Klasse mit der höchsten Wahrscheinlichkeit als Vorhersage aus.
